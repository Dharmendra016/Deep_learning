# Functional API

# Of type split

```python
from keras.layers import *
from keras.models import Model
```

```python
x = Input(shape=(3,))
```

```python
hidden1 = Dense(64, activation='relu')(x)
hidden2 = Dense(128, activation='relu')(hidden1)
```

```python
output1 = Dense(1, activation="linear")(hidden2)
output2 = Dense(2, activation='sigmoid')(hidden2)
```

```python
model = Model(inputs=x, outputs=[output1, output2])
```

```python
from keras.utils import plot_model
plot_model(model, show_shapes=True)
```

# Of type concatenation

```python
from keras.layers import *
from keras.models import Model
```

```python
inputA = Input(shape=(32,))
inputB = Input(shape=(128,))
```

```python
# pathA
x = Dense(128, activation='relu')(inputA)
x1 = Dense(64, activation='relu')(x)
```

```python
# pathB
y = Dense(64, activation='relu')(inputB)
y1 = Dense(32, activation='relu')(y)
y2 = Dense(16, activation='relu')(y1)
```

```python
combined = concatenate([x1,y2])
```

```python
z = Dense(8, activation='relu')(combined)
z1 = Dense(4, activation='relu')(z)
```

```python
model1 = Model(inputs=[inputA, inputB], outputs=z1)
```

```python
from keras.utils import plot_model
plot_model(model1, show_shapes=True)
```

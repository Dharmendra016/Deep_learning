```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4 , n_features =1 , n_informative=1 , n_targets=1 ,noise=80 , random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
#lets apply ols
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X , reg.predict(X), color='green')
```

```python
# lets try using gredient descent
# finding the value of y at some random value of b as we assume that we knew the value of m already i.e 78.35
# lets choose b as 5
y_pred = ((78.35 * X)+ 100).reshape(4)
y_pred
```

```python
#plot
plt.scatter(X,y)
plt.plot(X , reg.predict(X), color='green' , label="ols")
plt.plot(X, y_pred, color='red' , label="b->5")
plt.legend()
plt.show()
```

```python
m = 78.35
b = 100
slope = -2 * np.sum(y-m*X.ravel()-b)
slope
```

```python
stepSize = 0.1 * slope
stepSize
```

```python
b_new = 100 - stepSize
b_new
```

```python
#1st iteration
y_pred1 = ((78.35 * X)+ b_new).reshape(4)
y_pred1
```

```python
#plot
plt.scatter(X,y)
plt.plot(X , reg.predict(X), color='green' , label="ols")
plt.plot(X, y_pred, color='red' , label="b->5")
plt.plot(X , y_pred1 , color="yellow",label="new-b")
plt.legend()
plt.show()
```

```python
# 2nd iteration
slope = -2 * np.sum(y-m*X.ravel()-b_new)
slope
```

```python
stepSize = 0.1 * slope
stepSize
```

```python
b_new  = b_new - stepSize
b_new
```

```python
y_pred2 = ((78.35 * X)+ b_new).reshape(4)
y_pred2
```

```python
#plot
plt.scatter(X,y)
plt.plot(X , reg.predict(X), color='green' , label="ols")
plt.plot(X, y_pred1, color='red' , label="b->5")
plt.plot(X , y_pred2 , color="yellow",label="new-b")
plt.legend()
plt.show()
```

```python
#3rd iteration
slope = -2 * np.sum(y-78.35*X.ravel()-b_new)
slope
```

```python
stepSize = 0.1 * slope
stepSize
```

```python
b_new  = b_new - stepSize
b_new
```

```python
y_pred3 = ((78.35 * X)+ b_new).reshape(4)
y_pred3
```

```python
#plot
plt.scatter(X,y)
plt.plot(X , reg.predict(X), color='green' , label="ols")
plt.plot(X, y_pred2, color='red' , label="b->5")
plt.plot(X , y_pred3 , color="yellow",label="new-b")
plt.legend()
plt.show()
```

```python
b = -100
m = 78.35
lr = 0.1
epochs = 10

for i in range(epochs):
    loss_slope = -2 * np.sum(y - m*X.ravel() - b)
    b = b - (lr*loss_slope)

    y_pred = m * X + b
    plt.plot(X , y_pred)
plt.scatter(X,y)
```

```python
b = -100
m = 78.35
lr = 0.01 #small learning rate
epochs = 100 # increased epochs

for i in range(epochs):
    loss_slope = -2 * np.sum(y - m*X.ravel() - b)
    b = b - (lr*loss_slope)

    y_pred = m * X + b
    plt.plot(X , y_pred)
plt.scatter(X,y)
```
